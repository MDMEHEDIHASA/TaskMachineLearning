# Analyze this data and state a few facts about the data and Handling Missing Data

```python
def percent_missing(df):
    percent_nan = 100* df.isnull().sum() / len(df)
    percent_nan = percent_nan[percent_nan>0].sort_values()
    return percent_nan
```

# Apply Differtent Machine Learning Algorithm & find out train and test loss

```python
X = final_df.drop('Loan Approved',axis=1)
y = final_df['Loan Approved']
```

```python
X_test, X_eval, y_test, y_eval = train_test_split(X_test, y_test, test_size=0.5, random_state=101)
```

```python
len(X_test)
```

# APPlY KNN Classifier Algorithm 

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
test_error_rates = []
train_loss_rates = []

for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train,y_train) 
    
    y_pred_train = knn_model.predict(X_train)
    y_pred_test = knn_model.predict(X_eval)
    
    train_error = 1- accuracy_score(y_train,y_pred_train)
    test_error = 1 - accuracy_score(y_eval,y_pred_test)
    train_loss_rates.append(train_error)
    test_error_rates.append(test_error)
```

```python
plt.figure(figsize=(10,6),dpi=200)
plt.plot(range(1,30),train_loss_rates,label='Train Loss')
plt.legend()
plt.ylabel('Error Rate')
plt.xlabel("K Value")
```

```python
plt.figure(figsize=(10,6),dpi=200)
plt.plot(range(1,30),test_error_rates,label='Test Error')
plt.legend()
plt.ylabel('Error Rate')
plt.xlabel("K Value")
```

# Because the error rate is same for all test values. That's why we can choose any values of neighbors. I choose 5

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train,y_train)
y_pred_train = knn_model.predict(X_train)
y_pred_test = knn_model.predict(X_eval)
```

```python
print('Training Loss:',1-accuracy_score(y_train,y_pred_train))
print('Test Loss:',1-accuracy_score(y_eval,y_pred_test))
```

# Apply RandomForestClassifier Algorithm

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
errors = []
misclassifications = []

for n in range(1,64):
    rfc = RandomForestClassifier(class_weight='balanced',n_estimators=n,bootstrap=True,max_features='auto')
    rfc.fit(X_train,y_train)
    preds = rfc.predict(X_eval)
    err = 1 - accuracy_score(preds,y_eval)
    n_missed = np.sum(preds != y_test)
    errors.append(err)
    misclassifications.append(n_missed)
```

```python
plt.plot(range(1,64),errors)
```

```python
plt.plot(range(1,64),misclassifications)
```

```python
rfc = RandomForestClassifier(class_weight='balanced',n_estimators=10,bootstrap=True,max_features='auto')
rfc.fit(X_train,y_train)
y_pred_train = rfc.predict(X_train)
y_predict_test = rfc.predict(X_eval)
```
